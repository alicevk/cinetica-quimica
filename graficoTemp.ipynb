```python
# -------------------------------- Importações ------------------------------- #
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
# ---------------------------- Importando os dados --------------------------- #
dfConc = pd.read_csv('dados/Temp/T-Num=300.csv', header=None)

dfConc
```

```python
# ------------------------- Igualando as porcentagens ------------------------ #
eixo_x = np.array(dfConc[0], dtype=np.float128) * 1e-4

eixo_y = np.array(dfConc[1], dtype=np.float128)
```

```python
# ---------------------- Plotando os dados da simulação ---------------------- #
plt.figure(figsize=(8,5))

plt.title('Variação na temperatura do sistema')

plt.xlabel('Tempo (Frames)')
plt.ylabel('Temperatura (u.a.)')

plt.plot(eixo_x, eixo_y, color='red')

plt.grid()
plt.show() 
```
